# Coletar dados

```python
import yfinance as yf
```

```python
ticker = input("Digite o código da ação: ")
acoes = yf.Ticker(ticker).history("2y")
```

```python
acoes
```

```python
acoes.Close.plot()
```

# Tratamento dos dados

```python
# reinicializando o índice da tabela de dados
dados = acoes.reset_index()
dados
```

```python
treinamento = dados[["Date", "Close"]]
treinamento
```

```python
treinamento["Date"].dt.tz_localize(None)
```

```python
treinamento
```

```python
treinamento["Date"] = treinamento["Date"].dt.tz_localize(None)
treinamento
```

```python
treinamento.columns
```

```python
treinamento.columns = ["ds", "y"]
treinamento
```

# Treinando o modelo de dados

```python
!pip install yfinance prophet
```

```python
from prophet import Prophet
```

```python
modelo = Prophet()
```

```python
modelo
```

```python
modelo.fit(treinamento)
```

# Criar as previsões

```python
treinamento
```

```python
# adiciona 90 datas futuras
periodo = modelo.make_future_dataframe(90)
periodo
```

```python
previsoes = modelo.predict(periodo)
previsoes
```

# Visualizando as previsões

```python
from prophet.plot import plot_plotly
```

```python
plot_plotly(modelo, previsoes)
```
